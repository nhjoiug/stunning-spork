```python
import numpy as np
import matplotlib.pyplot as plt

vcharge = np.array([0.00, 3.57, 6.17, 7.83, 8.99, 9.73, 10.23, 10.57, 10.84, 10.99, 11.10, 11.19, 11.25])
t = np.array([0.0, 10.0, 20.0, 30.0, 40.0, 50.0, 60.0, 70.0, 80.0, 90.0, 100.0, 110.0, 120.0])

plt.figure(figsize=(15,10))
plt.plot(t, vcharge, color='purple', linestyle='--')
plt.scatter(t, vcharge, color='blue', marker='o')

plt.ylabel('Voltage (Volts)')
plt.xlabel('Time (seconds)')
plt.title('Voltage Increase of a Charging Capacitor in an RC circuit Over Time')
```
